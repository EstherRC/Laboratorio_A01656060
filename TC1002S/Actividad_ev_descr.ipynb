### User Followers An√°lisis

A continuaci√≥n se presentan los datos ob tenidos a partir de la variable num√©rica user_followers

```python
min = dfcovid["user_followers"].min()
max = dfcovid["user_followers"].max()
media = dfcovid["user_followers"].mean()
mediana = dfcovid["user_followers"].median()
moda = dfcovid["user_followers"].mode()
desviacion_estandar = np.std(dfcovid["user_followers"])

print(f"El rango es {min}-{max}\nLa media es {media}\nLa mediana es {mediana}\nLa moda es {moda}\nLa desviaci√≥n estandar es {desviacion_estandar}\n")
```
